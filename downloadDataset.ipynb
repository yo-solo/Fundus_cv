```python
from docx import Document

# 加载 DOCX 文件
docx_path = "2024年迎峰度夏6、7月用电分析V5.docx"  # 替换为您的文件路径
doc = Document(docx_path)

# 初始化变量
page_content = []  # 存储所有页的内容
current_page = []  # 存储当前页的内容

# 遍历段落，按分页符划分页面
for paragraph in doc.paragraphs:
    current_page.append(paragraph.text)  # 添加段落内容
    if "Page Break" in paragraph.text:  # 假设分页符为 "Page Break"
        page_content.append("\n".join(current_page))  # 保存当前页内容
        current_page = []  # 开始新的一页

# 添加最后一页（如果没有分页符结尾）
if current_page:
    page_content.append("\n".join(current_page))

# 打印前两页内容
print("第一页的内容：\n")
print(page_content[0] if len(page_content) > 0 else "第一页内容为空。")
print("\n第二页的内容：\n")
print(page_content[1] if len(page_content) > 1 else "第二页内容为空。")
```

```python
from docx import Document

def extract_text_and_tables(doc):
    """
    提取文档中的文本和表格内容，按顺序返回。
    """
    elements = []
    for element in doc.element.body:
        if element.tag.endswith('p'):  # 段落
            elements.append({'type': 'text', 'content': element.text})
        elif element.tag.endswith('tbl'):  # 表格
            table = []
            for row in element.iter():
                if row.tag.endswith('tr'):  # 表格行
                    table_row = [cell.text for cell in row.iter() if cell.tag.endswith('tc')]  # 获取每个单元格的文本
                    table.append(table_row)
            elements.append({'type': 'table', 'content': table})
    return elements

def print_page_content(page_elements):
    """
    打印页面内容，包括文本和表格。
    """
    for element in page_elements:
        if element['type'] == 'text':
            print(element['content'])
        elif element['type'] == 'table':
            print("\n[表格内容]")
            for row in element['content']:
                print("\t".join(row))  # 使用制表符分隔单元格内容

# 加载 DOCX 文件
docx_path = "冀北全社会用电基础负荷及降温.docx"  # 替换为文件路径
doc = Document(docx_path)

# 提取文本和表格内容
elements = extract_text_and_tables(doc)

# 分页逻辑（基于段落中的分页符）
pages = []
current_page = []
for element in elements:
    current_page.append(element)
    if element['type'] == 'text' and "Page Break" in element['content']:
        pages.append(current_page)
        current_page = []

# 添加最后一页（如果没有分页符结尾）
if current_page:
    pages.append(current_page)

# 打印前两页内容
print("第一页的内容：\n")
if len(pages) > 0:
    print_page_content(pages[0])
else:
    print("第一页内容为空。")

print("\n第二页的内容：\n")
if len(pages) > 1:
    print_page_content(pages[1])
else:
    print("第二页内容为空。")
```

```python
from PyPDF2 import PdfReader

# 打开 PDF 文件
pdf_path = "2024年迎峰度夏6、7月用电分析V5.pdf"  # 请将此路径替换为您的 PDF 文件路径

reader = PdfReader(pdf_path)

# 获取 PDF 文件的页数
num_pages = len(reader.pages)

# 遍历每一页并提取文本
for page_num in range(num_pages):
    page = reader.pages[page_num]
    text = page.extract_text()
    print(f"第 {page_num + 1} 页的内容：\n{text}\n")
```
